```python
# 2. What is an object in Python?
# Ans:
#     an object is an instance of a class. It is a fundamental concept of object-oriented programming (OOP) 
#     and represents a specific entity with its own attributes (data) and behaviors (methods).
#     In Python, almost everything is an object. Whether it's a built-in data type like integers, strings, lists,
#     or a user-defined class, they are all objects. Even functions, modules, and exceptions are objects in Python.
```

```python
    # 3. What is a class in Python?
# Ans:
#     In Python, a class is a blueprint or template that defines the structure and behavior of objects. It serves as a blueprint
#     for creating instances of objects, which are known as objects or instances of the class.
```

```python
# 4. What are attributes and methods in a class?
# Ans:
#     class defines the attributes (data) and methods (functions) that objects of the class will have. 
#     Attributes represent the state or characteristics of the objects, while methods define the behavior 
#     or actions that objects can perform.
```

```python
# 5. What is the difference between class variables and instance variables in Python?
# Ans:
#     The main difference between them lies in their scope and how they are accessed.

#     Class Variables:

#     Class variables are variables that are shared among all instances (objects) of a class. 
#     They are defined within the class but outside of any methods.
#     Class variables are associated with the class itself rather than with individual instances.
#     They are accessible by all instances of the class as well as the class itself.
#     Changes made to a class variable are reflected across all instances and the class.
#     Class variables are typically used to store data that is shared by all instances of the class.
    
    
#     Instance Variables:

#     Instance variables are variables that are unique to each instance of a class.
#     They are defined within the class methods or the class constructor (__init__ method).
#     Instance variables are specific to individual objects and hold data that varies between instances.
#     Each instance of a class has its own copy of instance variables, and changes made to these variables only affect
#     the specific instance.
#     Instance variables are accessed using the instance (object) name followed by the variable name.
#     They can have different values for each instance of the class.
```

```python
# 6. What is the purpose of the self parameter in Python class methods?
# Ans:
#     The self parameter in class methods represents the instance of the class. It is a convention to use the name self,
#     but any other valid identifier can be used as well.

#     When you define a class method, the first parameter is always self. 
#     This parameter is automatically passed to the method when it is called on an instance of the class.
#     By convention, self is used to refer to the instance itself within the method.

#     The purpose of the self parameter is to provide a way for the instance to refer to its own attributes and methods.
#     It allows accessing and modifying the instance variables and invoking other methods of the same instance.
```

```python
# 7. For a library management system, you have to design the "Book" class with OOP
# principles in mind. The “Book” class will have following attributes:
# a. title: Represents the title of the book.
# b. author: Represents the author(s) of the book.
# c. isbn: Represents the ISBN (International Standard Book Number) of the book.
# d. publication_year: Represents the year of publication of the book.
# e. available_copies: Represents the number of copies available for checkout.
# The class will also include the following methods:
# a. check_out(self): Decrements the available copies by one if there are copies
# available for checkout.
# b. return_book(self): Increments the available copies by one when a book is
# returned.
# c. display_book_info(self): Displays the information about the book, including its
# attributes and the number of available copies.

# creating class Book
class Book:
    def __init__(self, title,author,isbn,publication_year,available_copies):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.publication_year=publication_year
        self.available_copies=available_copies
    
# creating method check_out      
    def check_out(self, title):
        if self.available_copies>0:
            self.available_copies -= 1
            print(f"Book {self.title} checked out successfully")
        else:
            print(f"Sorry, '{self.title} by {self.author}' is not available")
# creating method return_book              
    def return_book(self, title):
        self.available_copies += 1
        print(f"Book {self.title} returned successfully")
# creating method display_book_info           
    def display_book_info(self, title):
        from textwrap import dedent
        print(dedent(f"""
            Book Name: {self.title}
            Book Author: {self.author}
            Book ISBN: {self.isbn}
            Book Publication: {self.publication_year}
            Available copies: {self.available_copies}"""))

# Creating a book instance
book1 = Book("Arthashastra","Kautilya", "1234", "1800", 10)
book2 = Book("Time Machine", "H.G. Wells", "7856875", "1950", 13)
# Checking out the book(s)
book1.check_out("Arthashastra")
book2.check_out("Time Machine")
# Returning the book(s)
book1.return_book("Arthashastra")
book2.return_book("Time Machine")
# Displaying the book information
book1.display_book_info("Arthashastra")
book2.display_book_info("Time Machine")
```

```python
# 8. For a ticket booking system, you have to design the "Ticket" class with OOP
# principles in mind. The “Ticket” class should have the following attributes:
# a. ticket_id: Represents the unique identifier for the ticket.
# b. event_name: Represents the name of the event.
# c. event_date: Represents the date of the event.
# d. venue: Represents the venue of the event.
# e. seat_number: Represents the seat number associated with the ticket.
# f. price: Represents the price of the ticket.
# g. is_reserved: Represents the reservation status of the ticket.
# The class also includes the following methods:
# a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
# b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
# reserved.
# c. display_ticket_info(self): Displays the information about the ticket, including its
# attributes and reservation status.

# creating class Ticket
class Ticket:
    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price):
        self.ticket_id=ticket_id
        self.event_name=event_name
        self.event_date=event_date
        self.venue=venue
        self.seat_number=seat_number
        self.price=price
        self.is_reserved=False
# creating reserve_ticket method to check if ticket is already reserved or not   
    def reserve_ticket(self):
        if not self.is_reserved:
            self.is_reserved = True
            print("Ticket reserved successfully.")
        else:
            print("Ticket is already reserved.")
            
            
# creating cancel_reservation method to check if reservation needs to be cancelled or not.         
    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False
            print(f"Ticket {self.ticket_id} reservation canceled successfully.")
        else:
            print("Ticket is not reserved.")

# creating display_ticket_info method to diplay the ticket information
    def display_ticket_info(self):
        from textwrap import dedent #dedent function is used to remove the indentation
        print(dedent(f"""
        Ticket Information:
        Ticket ID: {self.ticket_id}
        Event Name: {self.event_name}
        Event Date: {self.event_date}
        Venue: {self.venue}
        Seat Number: {self.seat_number}
        Price: {self.price}"""))
        print("Reservation Status:", "Reserved" if self.is_reserved else "Not Reserved")

# Creating a ticket instance
ticket1 = Ticket("T123", "Concert", "2023-07-15", "Music Hall", "A12", 100.00)

# Reserving the ticket
ticket1.reserve_ticket()

# Cancelling the ticket reservation
ticket1.cancel_reservation()

# Displaying the ticket information
ticket1.display_ticket_info()      
```

```python
# 9. You are creating a shopping cart for an e-commerce website. Using OOP to model
# the "ShoppingCart" functionality the class should contain following attributes and
# methods:
# a. items: Represents the list of items in the shopping cart.
# The class also includes the following methods:
# a. add_item(self, item): Adds an item to the shopping cart by appending it to the
# list of items.
# b. remove_item(self, item): Removes an item from the shopping cart if it exists in
# the list.
# c. view_cart(self): Displays the items currently present in the shopping cart.
# d. clear_cart(self): Clears all items from the shopping cart by reassigning an
# empty list to the items attribute.


# creating class ShoppingCart
class ShoppingCart:
    def __init__(self):
        self.items=[]
        
# creating method add_item to add an item to the shopping cart
    def add_item(self, item):
        self.items.append(item)
        print(f"Item '{item}' added to the shopping cart.")
        
# creating method remove_item to remove the items from teh shopping cart
    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print(f"Item '{item}' removed from the shopping cart.")
        else:
            print(f"Item '{item}' is not present in the shopping cart.")
            
# creating methpod view_cart to view the items present in the cart
    def view_cart(self):
        print("Shopping Cart:")
        if self.items:
            for item in self.items:
                print("- ", item)
        else:
            print("The shopping cart is empty.")
# creating method clear_Cart to remove all teh items from the cart
    def clear_cart(self):
        self.items = []
        print("Shopping cart cleared.")

        
item1 = ShoppingCart()

# Create a shopping cart instance
cart = ShoppingCart()

# Add items to the cart
cart.add_item("Self Discipline")
cart.add_item("Confidence")
cart.add_item("Self Belief")
cart.add_item("Procastination")

# View the items in the cart
cart.view_cart()

# Remove an item from the cart
cart.remove_item("Procastination")

# Clear the cart
cart.clear_cart()

# View the items in the cart after clearing
cart.view_cart()
```

```python
# 10. Imagine a school management system. You have to design the "Student" class using
# OOP concepts.The “Student” class has the following attributes:
# a. name: Represents the name of the student.
# b. age: Represents the age of the student.
# c. grade: Represents the grade or class of the student.
# d. student_id: Represents the unique identifier for the student.
# e. attendance: Represents the attendance record of the student.
# The class should also include the following methods:
# a. update_attendance(self, date, status): Updates the attendance record of the
# student for a given date with the provided status (e.g., present or absent).
# b. get_attendance(self): Returns the attendance record of the student.
# c. get_average_attendance(self): Calculates and returns the average
# attendance percentage of the student based on their attendance record.


# creating class Student
class Student:
    def __init__(self, name, age, grade, student_id):
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = {}
# creating method update_attendance
    def update_attendance(self, date, status):
        self.attendance[date] = status
# creating method get_attendance
    def get_attendance(self):
        return self.attendance
    
# creating method get_average_attendance
    def get_average_attendance(self):
        total_days = len(self.attendance)
        if total_days > 0:
            present_days = sum(status == "present" for status in self.attendance.values())
            attendance_percentage = (present_days / total_days) * 100
            return attendance_percentage
        else:
            return 0

# Create a student instance
student = Student("Prince Batra", 16, "10th Grade", "12345")

# Update attendance
student.update_attendance("2023-07-01", "present")
student.update_attendance("2023-07-02", "absent")
student.update_attendance("2023-07-03", "present")

# Get attendance record
attendance_record = student.get_attendance()
print(attendance_record)

# Get average attendance percentage
average_attendance = student.get_average_attendance()
print("Average Attendance:", average_attendance)
```
