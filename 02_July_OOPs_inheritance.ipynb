```python
# 1. Explain what inheritance is in object-oriented programming and why it is used.
# Ans:
#     Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class to inherit attributes
#     and behaviors from another class. It enables the creation of a hierarchy of classes, where subclasses inherit and 
#     extend the properties and functionalities of a superclass
    
#     Inheritance is used in OOP for the following purposes:
#     Code Reusability
#     Modelling Real-World Relationships
#     Polymorphism
#     Overriding and Extending Functionalities
    
```

```python
# 2. Discuss the concept of single inheritance and multiple inheritance, highlighting their
# differences and advantages.
# Ans:
#     Single Inheritance:

#     Single inheritance refers to the ability of a class to inherit from a single superclass or base class.
#     In single inheritance, a subclass extends and inherits the attributes and methods of a single superclass.
#     The subclass forms a parent-child relationship with the superclass, where the subclass inherits the characteristics
#     of the superclass and can add its own unique features.
#     It helps in organizing code and promotes code reuse, as subclasses can inherit and extend the functionality of a single superclass.
    
#     Multiple Inheritance:

#     Multiple inheritance refers to the ability of a class to inherit from multiple superclasses or base classes.
#     In multiple inheritance, a subclass can inherit attributes and behaviors from multiple superclasses, 
#     forming a more complex class hierarchy.
#     The subclass inherits characteristics from all the superclasses, allowing for combining and reusing code 
#     from multiple sources.
#     Multiple inheritance allows for modeling complex relationships and creating specialized classes that inherit properties
#     from multiple sources.
#     It provides flexibility and the ability to capture diverse features and behaviors from different classes.
    
#     Advantages of Single Inheritance:

#     Simplicity: Single inheritance provides a simpler and more straightforward class hierarchy, making code organization
#     and understanding easier.
#     Clarity: With single inheritance, there is a clear parent-child relationship between classes, making the structure 
#     and flow of the code more evident.
#     Reduced Complexity: Single inheritance reduces the chances of conflicts and ambiguity that may
#     arise in multiple inheritance scenarios.
    
#     Advantages of Multiple Inheritance:

#     Code Reusability: Multiple inheritance allows for combining and reusing code from multiple sources, promoting code reuse
#     and modularity.
#     Flexibility: Multiple inheritance provides flexibility to create specialized classes that inherit and integrate feature
#     from multiple superclasses, capturing diverse characteristics.
#     Modelling Complex Relationships: Multiple inheritance enables the modeling of complex relationships and interactions
#     between objects, reflecting real-world scenarios more accurately.
    
    
```

```python
# 3. Explain the terms "base class" and "derived class" in the context of inheritance.
# Ans:
#     Base Class:

#     A base class, also known as a superclass or parent class, it is the class from which other classes inherit attributes and behaviors.
#     It serves as the starting point or foundation for the inheritance hierarchy.
#     The base class defines common attributes, methods, and behavior that can be shared by multiple derived classes.
#     It provides a blueprint or template for the derived classes to inherit from and extend.

#     Derived Class:

#     A derived class, also known as a subclass or child class, it is a class that inherits attributes and behaviors from a base class.
#     It extends the base class by adding its own unique attributes, methods, or behaviors, or by overriding inherited methods with its own implementation.
#     The derived class forms a specialized version of the base class, tailored to specific requirements or characteristics.
#     It inherits all the attributes and methods of the base class and can access and utilize them directly.
#     A derived class can have its own additional attributes, methods, or behaviors that are specific to its specialization.
```

```python
# 4. What is the significance of the "protected" access modifier in inheritance? How does
# it differ from "private" and "public" modifiers?
# Ans:
#     the significance of the "protected" access modifier in Python lies in the naming convention (_attribute) to indicate that 
#     an attribute or method is intended for internal use within the class and its subclasses. While it can still be accessed 
#     from outside the class, it is considered non-public.
    
#     The "private" access convention (__attribute) is similar but provides stronger indication and name-mangling to stop direct
#     access from outside the class. Ultimately, the use of protected and private attributes and methods relies on the discipline 
#     and agreement of the developers to respect the intended visibility and encapsulation.
    
#     In Python, by default, all attributes and methods of a class are considered public, meaning they can be accessed and modified
#     from anywhere outside the class.
```

```python
# 5. What is the purpose of the "super" keyword in inheritance? Provide an example.
# Ans:
#     In Python, the super keyword is used to refer to the superclass or base class or Parent class from within a subclass.
#     It provides a way to invoke the methods and access the attributes of the superclass.
#     The primary purpose of the super keyword is to facilitate method overriding and enable the subclass to extend or customize the behavior inherited from the superclass. It allows the subclass
#     to invoke the superclass's methods and add its own additional functionality.
    
#     Example:
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start_engine(self):
        print("Engine started.")

    def stop_engine(self):
        print("Engine stopped.")


class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  # Calling the superclass's __init__ method
        self.model = model

    def start_engine(self):
        super().start_engine()  # Calling the superclass's start_engine method
        print("Car engine started.")

    def stop_engine(self):
        super().stop_engine()  # Calling the superclass's stop_engine method
        print("Car engine stopped.")


# Create an instance of the Car class
car = Car("Toyota", "Camry")

# Access attributes from the superclass
print(car.brand)  # Output: Toyota
print(car.model)  # Output: Camry

# Invoke methods from the superclass
car.start_engine()  # Output: Engine started. Car engine started.
car.stop_engine()  # Output: Engine stopped. Car engine stopped.

        

```

```python
# 6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
# Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
# attribute called "fuel_type". Implement appropriate methods in both classes.
# Ans:
    
class Vehicle:
    def __init__(self, make, model, year):
        self.make=make
        self.model=model
        self.year=year

    def start_engine(self):
        print("Engine started.")

    def stop_engine(self):
        print("Engine stopped.")


class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)  # Calling the superclass's __init__ method
        self.fuel_type = fuel_type

    def start_engine(self):
        super().start_engine()  # Calling the superclass's start_engine method
        print("Car engine started.")

    def stop_engine(self):
        super().stop_engine()  # Calling the superclass's stop_engine method
        print("Car engine stopped.")


# Create an instance of the Car class
car = Car("Toyota", "Corrola", 2000, "Diesel")

# Access attributes from the superclass
print(car.model)  # Output: Corrola

# Invoke methods from the superclass
car.start_engine()  # Output: Engine started. Car engine started.
car.stop_engine()  # Output: Engine stopped. Car engine stopped.

        

```

```python
# 7. Create a base class called "Employee" with attributes like "name" and "salary."
# Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
# attribute called "department" for the "Manager" class and "programming_language"
# for the "Developer" class.
# Ans:
# creating parent class Employee
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

# creating child class Manager
class Manager(Employee):
    def __init__(self,name, salary, department):
        super().__init__(name,salary)
        self.department=department
        
# creating child class Developer
class Developer(Employee):
    def __init__(self,name,salary, programming_language):
        super().__init__(name,salary)
        self.programming_language=programming_language
```

```python
# 8. Design a base class called "Shape" with attributes like "colour" and "border_width."
# Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
# specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
# the "Circle" class.
# Ans:
# creating class Shape
class Shape:
    def __init__(self, colour, border_width):
        self.colour = colour
        self.border_width = border_width

# creating class rectangle which inherits the properties of parent class Shape
class Rectangle(Shape):
    def __init__(self, colour, border_width, length, width):
        super().__init__(colour, border_width)
        self.length = length
        self.width = width

# creating class Circle which inherits the properties of parent class Shape
class Circle(Shape):
    def __init__(self, colour, border_width, radius):
        super().__init__(colour, border_width)
        self.radius = radius
```

```python
# 9. Create a base class called "Device" with attributes like "brand" and "model." Derive
# two classes, "Phone" and "Tablet," from "Device." Add specific attributes like
# "screen_size" for the "Phone" class and "battery_capacity" for the "Tablet" class.
# Ans:
# creating class Device
class Device:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

# creating class Phone which inherits the properties of parent class Device
class Phone(Device):
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size

# creating class Tablet which inherits the properties of parent class Device
class Tablet(Device):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
    
```

```python
# 10. Create a base class called "BankAccount" with attributes like "account_number" and
# "balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
# "BankAccount." Add specific methods like "calculate_interest" for the
# "SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.
# Ans:
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance


class SavingsAccount(BankAccount):
    def calculate_interest(self, interest_rate):
        interest = (self.balance * interest_rate) / 100
        self.balance += interest
        print("Interest calculated and added to the account.")


class CheckingAccount(BankAccount):
    def deduct_fees(self, fee_amount):
        if self.balance >= fee_amount:
            self.balance -= fee_amount
            print("Fees deducted from the account.")
        else:
            print("Insufficient balance to deduct fees.")


# Create instances of the derived classes
savings_account = SavingsAccount("123456789", 1000)
checking_account = CheckingAccount("987654321", 500)

# Access and print attributes
print("Savings Account Attributes:")
print("Account Number:", savings_account.account_number)
print("Balance:", savings_account.balance)

print("\nChecking Account Attributes:")
print("Account Number:", checking_account.account_number)
print("Balance:", checking_account.balance)

# Call specific methods
savings_account.calculate_interest(5)  # Calculate interest at 5%
checking_account.deduct_fees(20)  # Deduct fees of Rs. 20

# Print updated balances
print("\nUpdated Balances:")
print("Savings Account Balance:", savings_account.balance)
print("Checking Account Balance:", checking_account.balance)
```
